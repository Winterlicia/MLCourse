# This class teaches how to model regressions using matplotlib and numpy


## TOOLS
NumPy for scientific computing
matplotlib for plotting data

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

Data points:
Size (sqft thousand) | Price ($ thousand)
1.0                   300
2.0                   500
GOAL: Fit a linear regression model through these two data points and predict price for other houses

x-train is input variable (x-value of regression), y-train is target output variable (predicted y-value)

```python
x_data = [1.0, 2.0]
y_data = [300.0, 500.0]
x_train = np.array(x_data)
y_train = np.array(y_data)
print(f"x-train = {x_train}") # [1. 2.]
print(f"y-train = {y_train}") # [300. 500.]
```

## NUMBER OF TRAINING EXAMPLES (m)

m = number of training examples
numpy arrays have .shape parameter. 
x_train.shape() = returns a python tuple (array/list) with an entry for each dimension. 
x_train.shape[0] = length of array and number of examples

```python
print(f"x_train.shape: {x_train.shape}") # x_train.shape: (2,)
m = x_train.shape[0] # Also m = len(x_train)
print(f"Number of training examples: {m}") # Number of training examples: 2
```

## TRAINING EXAMPLE (x^i, y^i)

(x^i, y^i) denotes i'th training example
Therefore (x^0, y^0) = (1.0, 300.0) and (x^1, y^1) = (2.0, 500.0)

```python
i = 0
Xi = x_train[i]
Yi = y_train[i]
print(f"(x^({i}), y^({i})) = ({Xi}, {Yi})")
```

## PLOTTING DATA USING MATPLOTLIB

Plotting the two points using the scatter() function in matplotlib library
Function arguments marker and c show the points as red crosses (default is blue dots)

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.xlabel("Size in ft^2 (thousands)")
plt.ylabel("Price in $ (thousands)")
plt.show()
```

## MODEL FUNCTION

Model function for the linear regression Fwb(x^i) = wx^i + b
x = input value, b = y-intercept parameter, w = slope param
Have to choose the correct parameters that best suit the data 

```python
w = 200
b = 100

def computeModelOutput(x_train, slope, b):
    '''
    Computes the prediction of a linear model:
    ndarray (m,) = describes a Numpy n-dimensional array of shape (m,)

    Arguments:
    - x_train (ndarray (m,)): Data, number of m examples
    - w,b (scalar):           model parameters
    Returns:
    - y (ndarray (m,)): Target values set of data
    '''
    m = x_train.shape[0] # Number of training examples, allows us to compute the number of values the model has in its dataset
    modelFwb = np.zeros(m) # 1D numpy array with 'm' entries -- in this case entries will represent the model's data points
    for i in range(m):
        modelFwb[i] = slope * x_train[i] + b # Set the model's values

    return modelFwb

# Plot the model using the function to calculate
plt.plot(x_train, computeModelOutput(x_train, w, b,), c='b', label = "Prediction")
plt.scatter(x_train, y_train, marker='x', c='r', label = 'Actual Values')
plt.title("Housing Prices")
plt.xlabel("Size in ft^2 (thousands)")
plt.ylabel("Price in $ (thousands)")
plt.legend()
plt.show()
```
